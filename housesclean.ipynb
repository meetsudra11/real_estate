```python
import numpy as np
import pandas as pd
import re
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('independent-house.csv')
df.head()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df.drop(columns=['link','property_id'], inplace=True)
```

# price_per_sqrt

```python
df.rename(columns={'rate':'price_per_sqft'}, inplace=True)
```

```python
df['price_per_sqft']
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
df.head()
```

# society column

```python
df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

- some society names have ratings as well in their cols 
- some society names are repeating just because of upper and lower case 

```python
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
# .strip() is used to remove any leading n trailing spaces
```

```python
df['society'] = df['society'].str.replace('nan','independent')
```

# price column

```python
df['price'].value_counts()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

# bedroom, bathroom, balcony column

```python
df['bedRoom'].value_counts()
```

```python
df['bathroom'].value_counts()
```

```python
df['balcony'].value_counts()
```

```python
print(df['bathroom'].isnull().sum())
print(df['bedRoom'].isnull().sum())
print(df['balcony'].isnull().sum())
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
a = ['bathroom','bedRoom']
for i in a:
    df[i] = df[i].str.split(' ').str.get(0).astype('int')
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df.head()
```

# additionalRoom column

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'].fillna('Not available', inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.head()
```

# noOfFloor column

```python
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
df['floorNum'].value_counts()
```

```python
df['floorNum'].isnull().sum()
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0)
```

# Facing column

```python
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df['facing'].value_counts()
```

# Inserting col 'property_type' 

```python
df.insert(loc=1,column='property_type',value='house')
```

```python
df['area'] = round((df['price']*10000000)/df['price_per_sqft'])
```

```python
df.info()
```

```python
df.to_csv('house_cleaned.csv',index=False)
```
